```python
import pandas as pd
import numpy as np
import pickle
import os
```

```python
df=pd.read_csv('movies_1.csv')
```

```python
df=df.drop_duplicates(subset=['name_name','name_url'])
df.info()
```

```python
df.columns=df.columns.str.replace('name_','')
df=df.drop(columns=['ratings'])
df.sample(10)
df['text']=np.where(df['text']==pd.NA,'missing',df['text'])
```

```python
df=df.fillna(value='missing')
df.index=np.arange(0,811)
```

```python
title_number=np.zeros(df.shape[0])
df.info()
```

```python
tracks=df['url'].str.replace('https://www.imdb.com/title/','')
tracks=tracks.tolist()
```

##All the extraction process is done above this so run upto this majale:)


```python
genre=pd.read_csv('getting_datas/genre_list.csv',sep='\t',header=None).T
casts=pd.read_csv('getting_datas/cast.csv',sep='\t' ,header=None).T
```

```python
director=pd.read_csv('getting_datas/directors.csv',sep='\t',header=None ).T
director.sample(10)
```

```python



#director preprocessing
director_s=director.iloc[:,0]
director_s=director_s.str.replace(' ','')
director_s=director_s.str.lower()
director_s=director_s[:-1]
director_s.shape
```

```python
#genre preprocessing
genre_s=genre.iloc[:,0]
genre_s=genre_s.str.replace('[',"").str.replace(']',"").str.replace("'","")
genre_s=genre_s.str.lower()
genre_s=genre_s[:-1]
```

```python
#cast preprocessing
cast_s=casts.iloc[:,0]
cast_s=cast_s.str.replace('[','').str.replace(']','').str.lower()
cast_s=cast_s[:-1]
```

```python
#concatinating the df and the three seires
df['cast']=cast_s
df['genre']=genre_s
df['director']=director_s
```

```python
df.head()
```

```python
#convering the number in the name into proper form
filler=r"^\d+\.\s+"

text=df['name'].str.replace( filler,"", regex=True)

df['name']=text
```

```python
df.to_csv('movies_final.csv',index=True)
```

```python
cast_list=df['cast'].str.split(",")
cast_list=cast_list.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
genre_list=df['genre'].str.split(',')
director_list=df['director'].str.split(" ")
genre_list
```

```python
text_list=df['text'].str.split(" ")
x=text_list+genre_list+director_list+cast_list
to_vectorize_df=pd.DataFrame(x,columns=['desc'])
```

```python
to_vectorize_df['movie_name']=df['name']
```

```python
to_vectorize_df['track_num']=title_num
```

```python


to_vectorize_df['desc']=[str(item).replace(','," ").replace("'","") for item in to_vectorize_df['desc']]
to_vectorize_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000,stop_words='english')
vectors=cv.fit_transform(to_vectorize_df['desc']).toarray()
```

```python
print(cv.get_feature_names_out())
```

```python

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))


    return " ".join(y)
```

```python
to_vectorize_df['desc']=to_vectorize_df['desc'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie_name):
    movie_index=to_vectorize_df[to_vectorize_df['movie_name']==movie_name].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(to_vectorize_df.iloc[i[0]].movie_name)


```

```python
movie_to_test=to_vectorize_df.loc[np.random.randint(0,810),'movie_name']
print(movie_to_test)
```

```python
recommend('Kabaddi')
```

```python
pickle.dump(similarity, open('similarity_matrix.pkl','wb'))
```
